```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = col_selecionados2.drop('DESCR_TIPO_VEICULO', 1)
y = col_selecionados2['DESCR_TIPO_VEICULO']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.2)

modelo = LinearRegression()
modelo.fit(X_train,y_train)

predicao = modelo.predict(X_test,y_test)
```

```python

from sklearn.metrics import mean_squared_error,mean_absolute_error

MAE = mean_absolute_error(y_test,predicao)
RMSE = np.sqrt(mean_squared_error(y_test,predicao))


print(MAE)
print(RMSE)
```
